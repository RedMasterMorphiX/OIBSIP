Calculating the cosmic vibes: üåå Average, üöÄ Highest, and üå† Lowest Unemployment Rate

```python
# Channeling the cosmic currents to unveil the societal pulse
universe_pulse = cosmic_data.groupby('State')['Estimated Unemployment Rate (%)'].mean()

# Synchronizing with celestial rhythms to uncover the extremes
zenith_state = universe_pulse.idxmax()
zenith_rate = universe_pulse.max()

nadir_state = universe_pulse.idxmin()
nadir_rate = universe_pulse.min()

# Decoding the cosmic messages to reveal the cosmic dance
print("State with the highest cosmic unemployment rate:", zenith_state)
print("Highest cosmic unemployment rate:", zenith_rate)
print("State with the lowest cosmic unemployment rate:", nadir_state)
print("Lowest cosmic unemployment rate:", nadir_rate)

# Visualizing the cosmic ebbs and flows with a different color palette
sns.set_palette("viridis")

plt.figure(figsize=(12, 6))
universe_pulse.sort_values(ascending=False).plot(kind='bar')
plt.title("Cosmic Pulse: Average Unemployment Rate by State")
plt.xlabel("State")
plt.ylabel("Average Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.show()
```

```python
# üéØ Mapping the cosmic interplay between Unemployment Rate and Labor Participation Rate üåå

# Setting the palette to a spectrum of cosmic hues
sns.set_palette("plasma")

# Creating the cosmic canvas
plt.figure(figsize=(10, 6))

# Plotting the cosmic journey
sns.scatterplot(x='Estimated Unemployment Rate (%)', y='Estimated Labour Participation Rate (%)', data=cosmic_data, hue='State', palette='plasma', legend='full', style='State', markers=True)

# Adding cosmic annotations
plt.title('Unemployment Rate vs. Labor Participation Rate')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.legend(title='State', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.grid(True)

# Presenting the cosmic spectacle
plt.show()
```

```python
# Extracting relevant numerical columns for correlation analysis
correlation_columns = cosmic_data[['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']]

# Calculating the correlation matrix
correlation_matrix = correlation_columns.corr()

# Creating the cosmic canvas
plt.figure(figsize=(8, 6))

# Plotting the cosmic connections with a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='magma')

# Adding cosmic annotations
plt.title('Correlation Matrix', fontsize=16, fontweight='bold')

# Presenting the cosmic insights
plt.show()
```

```python
# üé® Channeling cosmic energies into a vibrant depiction of Unemployment Rate by Frequency and State üåà

# Defining custom colors for each state
colors = sns.color_palette("husl", n_colors=len(cosmic_data['State'].unique()))

# Grouping the data by 'Frequency' and 'State' and calculating the mean of 'Estimated Unemployment Rate (%)'
grouped_data = cosmic_data.groupby(['Frequency', 'State'])['Estimated Unemployment Rate (%)'].mean().unstack()

# Creating the cosmic canvas
plt.figure(figsize=(12, 6))

# Plotting the cosmic saga with a bar plot
grouped_data.plot(kind='bar', stacked=True, color=colors)

# Adding cosmic annotations
plt.xlabel('Frequency (2019 and 2020)', fontsize=12)
plt.ylabel('Estimated Unemployment Rate (%)', fontsize=12)
plt.title('Unemployment Rate by Frequency and State', fontsize=14)
plt.legend(title='State', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.grid(True)

# Presenting the cosmic spectacle
plt.show()
```

üåç Embarking on a cosmic analysis of the entire country üöÄ

```python
# üìÖ Unraveling the cosmic chronology: Converting date strings to datetime objects üï∞Ô∏è
cosmic_data['Date'] = pd.to_datetime(cosmic_data['Date'].str.strip(), format='%d-%m-%Y')
```

```python
# üõ∏ Aligning the cosmic continuum: Setting the date as the index for seamless exploration üåå
cosmic_data.set_index('Date', inplace=True)
cosmic_data.head()
```

```python

# üöÄ Launching into cosmic abstraction: Shedding the 'State' column to focus on numeric data üåå
cosmic_data_numeric = cosmic_data.drop(columns=['State','Frequency','Area'])

# üìä Charting the cosmic averages: Grouping by date to calculate the mean of all numeric columns üìà
cosmic_data_numeric_avg = cosmic_data_numeric.groupby('Date').mean()

cosmic_data_numeric_avg
```

```python

# üìà Tracing the cosmic flux: Mapping the Estimated Unemployment Rate over time üåå
plt.figure(figsize=(12, 6))

# Plotting the cosmic journey with custom style and colors
plt.plot(cosmic_data_numeric_avg.index, cosmic_data_numeric_avg['Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='magenta')

# Adding cosmic annotations
plt.title('Estimated Unemployment Rate Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.grid(True)

# Presenting the cosmic spectacle
plt.show()
```

```python


# üìä Unveiling the cosmic spectrum: Histogram and Density Plot of Estimated Unemployment Rate üåå

# Setting the cosmic canvas for the Histogram
plt.figure(figsize=(10, 6))
plt.hist(cosmic_data_numeric_avg['Estimated Unemployment Rate (%)'], bins=20, density=True, alpha=0.7, color='purple')
plt.title('Histogram of Estimated Unemployment Rate', fontsize=14)
plt.xlabel('Unemployment Rate (%)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()

# Setting the cosmic canvas for the Density Plot
plt.figure(figsize=(10, 6))
cosmic_data_numeric_avg['Estimated Unemployment Rate (%)'].plot(kind='kde', color='purple')
plt.title('Density Plot of Estimated Unemployment Rate', fontsize=14)
plt.xlabel('Unemployment Rate (%)', fontsize=12)
plt.grid(True)
plt.show()
```
