```python

# Immerse the selected features in the crucible of scaling
selected_features = ["Car_Name", "Year", "Selling_Price", "Present_Price", "Driven_kms"]

# Invoke the power of the Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Transmute the selected features
df[selected_features] = scaler.fit_transform(df[selected_features])

# The transformation is complete, behold the scaled dataframe!
```

Gradient Boost regressor

```python

# Harness the forces of division and ensemble
from sklearn.model_selection import train_test_split  # Unleash the division
from sklearn.ensemble import GradientBoostingRegressor  # Invoke the ensemble

# Unleash the division of features and target
X = df.drop(columns="Selling_Price", axis=1)  # Gather the features
Y = df["Selling_Price"]  # Capture the target

# Initiate the split into training and testing realms
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)  # Engage the division with randomness
```

```python

# Forge the Regressor of Gradient Boosting with incantations of hyperparameters
gb_regressor = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)  # Craft the Regressor
gb_regressor.fit(x_train, y_train)  # Empower the Regressor with the wisdom of the training data

# Unleash the predictions upon the testing realm
pred1 = gb_regressor.predict(x_test)  # Divine the predictions with the power of Gradient Boosting
```

```python

# Calculate the mystical metrics of prediction accuracy
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score  # Unleash the metrics of evaluation
import numpy as np  # Invoke the powers of numerical manipulation

# Enchant the realm with the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, pred1)
print("Mean Absolute Error (MAE):", mae)
print("==============================================================================")

# Conjure the forces of the Mean Squared Error (MSE)
mse = mean_squared_error(y_test, pred1)
print("Mean Squared Error (MSE):", mse)
print("==============================================================================")

# Invoke the spirits of the R-squared (R2) Score
r2 = r2_score(y_test, pred1)
print("R-squared (R2) Score:", r2)
print("==============================================================================")

# Unveil the secrets of the Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)  # Transmute the truth and predictions
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100  # Invoke the spirits of percentage error

mape = mean_absolute_percentage_error(y_test, pred1)
print("Mean Absolute Percentage Error (MAPE):", mape)
```

```python

# Unleash the visualization of Actual vs Predicted Selling Price with a touch of elegance
plt.figure(figsize=(12, 8))  # Craft a canvas of boundless possibility
sns.regplot(x=y_test, y=pred1, color="purple", marker="*", scatter_kws={"s": 100}, line_kws={"linewidth": 2, "color": "navy"}, label="Best Fit Line")  # Illuminate the path with the radiance of stars
plt.title('Actual vs Predicted Selling Price', fontsize=20, fontweight='bold')  # Engrave the title with the grandeur of predictions
plt.xlabel('Actual Test Values', fontsize=16, fontweight='bold')  # Label the x-axis with the essence of truth
plt.ylabel('Predicted Test Values', fontsize=16, fontweight='bold')  # Label the y-axis with the vision of prediction
plt.xticks(fontsize=14)  # Engrave the x-axis ticks with the wisdom of truth
plt.yticks(fontsize=14)  # Engrave the y-axis ticks with the insight of prediction
plt.legend(fontsize=14, loc="upper left")  # Embrace the legend with the lore of prediction
plt.grid(True, linestyle='--', alpha=0.5)  # Infuse the visualization with the structure of insight
sns.despine()  # Strip away unnecessary spines
plt.show()  # Reveal the visualization of Actual vs Predicted Selling Price
```
