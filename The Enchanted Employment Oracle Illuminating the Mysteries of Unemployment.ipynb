```python
from sklearn.model_selection import train_test_split

class DataShuffler:
    def __init__(self, data, target, test_size=0.2, random_state=None):
                # Gearing up the Data Shuffler with mystical data and targets

        self.data = data
        self.target = target
        self.test_size = test_size
        self.random_state = random_state

    def shuffle_data(self):
                # Unleashing the cosmic forces of train-test splitting

        X_train, X_test, y_train, y_test = train_test_split(self.data, self.target, 
                                                            test_size=self.test_size, 
                                                            random_state=self.random_state)
                # Channeling the ancient ritual of data shuffling

        shuffled_indices = np.random.permutation(len(X_train))
        X_train, y_train = X_train.iloc[shuffled_indices], y_train.iloc[shuffled_indices]
                # Proclaiming the success of the shuffling ritual

        print("Data shuffled successfully using the Data Shuffler algorithm!")
                # Presenting the enchanted train-test split to the mortal realm

        return X_train, X_test, y_train, y_test

# Preparing the data shuffler with magical parameters

data_shuffler = DataShuffler(X, y, test_size=0.2, random_state=42)

# Unleashing the powers of the Data Shuffler to shuffle the data

X_train, X_test, y_train, y_test = data_shuffler.shuffle_data()
```
